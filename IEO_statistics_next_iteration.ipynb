```python
import pandas as pd
import numpy as np
%config IPCompleter.greedy=True
```

```python
ieos = pd.read_csv('view_crowdsale_manager_crowdsale1.csv')
ieos = ieos[['__create_datetime', 'id', 'name', 'goal', 'started', 'closed']]
ieos.rename({'__create_datetime': 'create_datetime_crowdsale', 'id' : 'crowdsale_id', 'goal':'goal_usdt'}, axis = 1, inplace = True)
ieos = ieos[(ieos['name'] != 'TEST') & (ieos['name'] != 'DAO')]
ieos
```

```python
deals = pd.read_csv('view_crowdsale_manager_deal_latest.csv')
deals = deals[['__create_datetime', 'crowdsale_id', 'user', 'purchase_currency','cost']]
deals.rename({'__create_datetime':'create_datetime_deal'}, axis =1 , inplace = True )
deals
startups = pd.merge(ieos, deals, on = 'crowdsale_id', how = 'left')
currency = pd.read_csv('view_asset_manager_currency.csv')
del currency['#'] 
currency = currency.rename(columns = {'id': 'purchase_currency'})
startups = pd.merge(startups,currency, on = 'purchase_currency') 
startups
exchange_rate = pd.read_csv('exchange_rate_usdt.csv')
exchange_rate = exchange_rate.rename(columns = {'tag_name':'tag'})
startups = pd.merge(startups, exchange_rate, on = 'tag')
startups['total_usdt'] = startups['cost'] * startups['price_in_usdt']
startups
```

```python

from datetime import datetime, timedelta
def calc_data(row):
    date_1 = datetime.strptime(row['started'], '%Y-%m-%d %H:%M:%S.%f')
    date_2 = datetime.strptime(row['closed'], '%Y-%m-%d %H:%M:%S.%f')
    timedelta = date_2 - date_1
    return timedelta.days
#startups['closed'].fillna('2020-11-17 20:25:10.000000' ,inplace = True)
#startups['started'].fillna('2020-10-26 16:00:46.968870' ,inplace = True)
startups['period_IEO_in_days'] = startups.apply(calc_data, axis=1)
startups
```

```python
df = startups[['name', 'total_usdt', 'goal_usdt', 'user', 'period_IEO_in_days']]
df1 = startups.groupby('name', as_index = False)['total_usdt'].sum()
df2 = df.groupby('name', as_index= False).agg({'user': 'nunique'})
df3 = pd.merge(df2, df1, on = 'name', how = 'inner')
df5 = startups[['name', 'total_usdt', 'goal_usdt','period_IEO_in_days']]
df4 = df5.drop_duplicates(subset = ['name'])
final = df3.merge(df4, on = 'name', how = 'inner')
del final['total_usdt_y']
final.rename({'total_usdt_x':'total_usdt'}, axis = 1, inplace=True)
final['success'] = final['total_usdt'] > final ['goal_usdt']
final
```

```python
final = final.rename({"user":"amount_of_buyers"},axis = 1)
final.to_excel('final.xlsx')
```

```python
def round_2(num):
    return float("%.2f" % num)

final['total_usdt'] = final['total_usdt'].apply(round_2)

def comma_3(num):
    a = format(num, ",d")
    return a

final['goal_usdt'] = final['goal_usdt'].apply(comma_3)

with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    display(final)
```
